```python
import re


COMMENTS = re.compile(r'''
    (//[^\n]*(?:\n|$))    # Everything between // and the end of the line/file
    |                     # or
    (/\*.*?\*/)           # Everything between /* and */
''', re.VERBOSE)


def remove_comments(content):
    return COMMENTS.sub('\n', content)
```

```python
def remove_comment(content):
    index = 0
    comment_line_inside = False
    comment_block_level = 0
    result = []
    while index < len(content):
        if content[index] == '/' and index + 1 < len(content) and content[index+1] == '*':
            comment_block_level += 1
        elif content[index] == '/' and content[index-1] == '*':
            comment_block_level -= 1
        elif content[index] == '/' and index + 1 < len(content) and content[index + 1] == '/':
            comment_line_inside = True
        elif content[index] == '\n' and comment_line_inside == True:
            comment_line_inside = False
        elif not comment_line_inside and comment_block_level == 0:
            result.append(content[index])
        index += 1

    return ''.join(result)
```

```python
def remove_comments(filename1, filename2,lang):
    """ Remove all comments beginning with # from filename1 and writes
    the result to filename2
    """

    with open(filename1, 'r') as f:
        lines = f.readlines()

    with open(filename2, 'w') as f:
        if lang == '*.py':
            for line in lines:
                # Keep the Shebang line
                if line[0:2] == "#!":
                    f.writelines(line)
                # Also keep existing empty lines
                elif not line.strip():
                    f.writelines(line)
                # But remove comments from other lines
                else:
                    line = line.split('#')
                    stripped_string = line[0].rstrip()
                    # Write the line only if the comment was after the code.
                    # Discard lines that only contain comments.
                    if stripped_string:
                        f.writelines(stripped_string)
                        f.writelines('\n')
        else:
            text = ''
            for i in lines:
                text += i
            #COMMENTS = re.compile(r'''(//[^\n]*(?:\n|$))|(/\*.*?\*/)''',re.VERBOSE)
            out = re.sub('//.*?\n|/\*.*?\*/', '', text, flags=re.S)
            f.writelines(out)
```
