```python
import re
import os
import glob
```

```python
def remove_comments(filename1, filename2,lang):
    """ Remove all comments beginning with # from filename1 and writes
    the result to filename2
    """

    with open(filename1, 'r') as f:
        lines = f.readlines()

    with open(filename2, 'w') as f:
        if lang == '*.py':
            for line in lines:
                # Keep the Shebang line
                if line[0:2] == "#!":
                    f.writelines(line)
                # Also keep existing empty lines
                elif not line.strip():
                    f.writelines(line)
                # But remove comments from other lines
                else:
                    line = line.split('#')
                    stripped_string = line[0].rstrip()
                    # Write the line only if the comment was after the code.
                    # Discard lines that only contain comments.
                    if stripped_string:
                        f.writelines(stripped_string)
                        f.writelines('\n')
        else:
            text = ''
            for i in lines:
                text += i
            #COMMENTS = re.compile(r'''(//[^\n]*(?:\n|$))|(/\*.*?\*/)''',re.VERBOSE)
            out = re.sub('//.*?\n|/\*.*?\*/', '', text, flags=re.S)
            f.writelines(out)
```

```python
for i in ['*.py','*.c','*.java']:
    Codeset = 'C:/Users/nikhi/OneDrive/Documents/gitUploads/Frontend/dataset/v1'    #Path to you folder with files
    for code_path in sorted(glob.glob(os.path.join(Codeset, i))):
        code_path = code_path.replace("/","\\")
        if i == '*.py':
            temp = 'temp.py'
        elif i == '*.c':
            temp = 'temp.c'
        else:
            temp = 'temp.java'
        filename2= temp
        remove_comments(code_path,filename2,i)
        f = open(filename2,'r')
        print(f.read())
```
